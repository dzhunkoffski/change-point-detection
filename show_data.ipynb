# 000

```python
import numpy as np
from synth_ts.synth_timeserie_generator import find_change_points, demonstrate

STEP_SIZE = 0.05
MIN_STABILITY_DURATION = 1
ID = '000'
TS_PARAMS = {
    'x0': np.pi,
    'lambd': 125,
    'n_steps': 1500000,
    'alpha': 1.75,
    'D': 0.2
}

with open('timeserie/timeserie_000.npy', 'rb') as f:
    timeserie = np.load(f)
timeserie = timeserie * np.pi

cp_ixs = find_change_points(timeserie, MIN_STABILITY_DURATION)
timeserie = demonstrate(TS_PARAMS, STEP_SIZE, timeserie, cp_ixs, max_winlen=1500, bins=50, subseq_len=250)
```
